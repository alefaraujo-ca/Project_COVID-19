```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
cases = pd.read_csv('../Resources/Data/COVID19.csv')
```

```python
country = pd.read_csv('../Resources/Data/country/country.csv')
```

```python
cases = cases.rename(columns={'ISO_CountryID':'Country_Id'})
```

```python
cases.columns
```

```python
country.columns
```

```python
country = country[['ID', 'Country_Id', 'Name', 'Lat', 'Lon',
       'Population', 'Density(P/Km2)', 'Land Area(Km2)',
       'Fert.Rate', 'Med.Age', 'UrbanPop', 'WorldShare']]
```

```python
cases.info()
```

```python
results = pd.merge(cases, country, on='Country_Id', how='left')
```

```python
results.info()
```

```python
results.columns
```

```python
results = results[['Population', 'Density(P/Km2)', 'Land Area(Km2)',
       'Med.Age', 'UrbanPop', 'Confirmed', 'Deaths', 'Recovered']]
```
